## Worldwide Governance Indicators

Les informations concernant la définition des indicateurs sont les suivantes :

- **`CC` : Control of Corruption**
    - Control of Corruption captures perceptions of the extent to which public power is exercised for private gain, including both petty and grand forms of corruption, as well as capture" of the state by elites and private interests. Estimate gives the country's score on the aggregate indicator
- **`GE` : Government Effectiveness**
    - Government Effectiveness captures perceptions of the quality of public services, the quality of the civil service and the degree of its independence from political pressures, the quality of policy formulation and implementation, and the credibility of the government's commitment to such policies. Estimate gives the country's score on the aggregate indicator, in units of a standard normal distribution, i.e. ranging from approximately -2.5 to 2.5.	
- **`PV` :Political Stability and Absence of Violence/Terrorism**

Estimate gives the country's score on the aggregate indicator, in units of a standard normal distribution, i.e. ranging from approximately -2.5 to 2.5.
- **`RQ` : Regulatory Quality**
    - Regulatory Quality captures perceptions of the ability of the government to formulate and implement sound policies and regulations that permit and promote private sector development. Estimate gives the country's score on the aggregate indicator, in units of a standard normal distribution, i.e. ranging from approximately -2.5 to 2.5.
- **`RL` : Rule of Law**
    - Rule of Law captures perceptions of the extent to which agents have confidence in and abide by the rules of society, and in particular the quality of contract enforcement, property rights, the police, and the courts, as well as the likelihood of crime and violence. Estimate gives the country's score on the aggregate indicator, in units of a standard normal distribution, i.e. ranging from approximately -2.5 to 2.5.
- **`VA` : Voice and Accountability**
    - Voice and Accountability captures perceptions of the extent to which a country's citizens are able to participate in selecting their government, as well as freedom of expression, freedom of association, and a free media. Estimate gives the country's score on the aggregate indicator, in units of a standard normal distribution, i.e. ranging from approximately -2.5 to 2.5.

- Utiliser une CAH (avec Ward) pour chercher un nombre de classes intéressants, et récupérer la partition proposée par la CAH avec ce nombre de classes

```python
import pandas as pd

# Charger les données
data = pd.read_csv('wgi2019.csv')

selected_columns = ['Control of Corruption', 'Voice and Accountability', 'Political Stability and Absence of Violence/Terrorism', 'Government Effectiveness', 'Regulatory Quality', 'Rule of Law']
X = data[selected_columns]

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import ward, dendrogram  # Importer la fonction dendrogram

# Remplacez les valeurs non finies par la moyenne de la colonne respective
X = X.fillna(X.mean())

# Calculer les liaisons avec Ward
linkage_matrix = ward(X)

# Afficher le dendrogramme pour aider à choisir le nombre de classes optimal
plt.figure(figsize=(12, 8))
plt.title('Dendrogramme CAH')
dendrogram(linkage_matrix)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, ward

# Charger les données
data = pd.read_csv('wgi2019.csv')

selected_columns = ['Control of Corruption', 'Voice and Accountability', 'Political Stability and Absence of Violence/Terrorism', 'Government Effectiveness', 'Regulatory Quality', 'Rule of Law']
X = data[selected_columns]

# Remplacez les valeurs non finies par la moyenne de la colonne respective
X = X.fillna(X.mean())

# Calculer les liaisons avec Ward
linkage_matrix = ward(X)

# Afficher le dendrogramme pour aider à choisir le nombre de classes optimal
plt.figure(figsize=(12, 8))
plt.title('Dendrogramme CAH')
dendrogram(linkage_matrix)  
plt.show()

# Choisissez un nombre de clusters
n_clusters = 4  # Choisissez le nombre de clusters approprié

# Créer un objet AgglomerativeClustering avec le nombre de clusters choisi
clustering = AgglomerativeClustering(n_clusters=n_clusters)

# Ajuster le modèle et obtenir les étiquettes des clusters
cluster_labels = clustering.fit_predict(X)

# Afficher les étiquettes des clusters pour chaque échantillon
print("Étiquettes des clusters :")
print(cluster_labels)
```

- Chercher une partition avec $k$-means, en cherchant le nombre de classes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Charger les données
data = pd.read_csv('wgi2019.csv')

# Sélection des colonnes
selected_columns = ['Control of Corruption', 'Voice and Accountability', 'Political Stability and Absence of Violence/Terrorism', 'Government Effectiveness', 'Regulatory Quality', 'Rule of Law']
X = data[selected_columns]

# Remplacer les valeurs non finies par la moyenne de la colonne respective
X.fillna(X.mean(), inplace=True)

# Normalisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Recherche du nombre optimal de clusters (méthode Elbow)
inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Tracer le graphique Elbow
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.title('Méthode Elbow pour choisir le nombre de clusters (k)')
plt.show()

# Choix du nombre de clusters (k)
k_optimal = 4  # Remplacez par le nombre de clusters choisi

# Appliquer K-means avec le nombre optimal de clusters
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Afficher les étiquettes des clusters pour chaque échantillon
print("Étiquettes des clusters :")
print(cluster_labels)
```

- Comparer les deux partitions obtenues (CAH et $k$-means) :
    - En comparant les centres des classes
    - ou autres critères

```python
from sklearn.metrics import silhouette_score, adjusted_rand_score, pairwise_distances
from sklearn.preprocessing import StandardScaler

# Standardiser les données pour la comparaison des centres
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Effectuer K-means avec le nombre optimal de clusters
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Calculer les centres des classes pour K-means
kmeans_centers = kmeans.cluster_centers_

# Comparaison des centres des classes entre CAH et K-means
cah_centers = X.groupby(cluster_labels).mean().values

# Calculer le Silhouette Score pour les deux partitions
silhouette_cah = silhouette_score(X_scaled, cluster_labels)
silhouette_kmeans = silhouette_score(X_scaled, kmeans_labels)

# Calculer l'indice de similarité (ajusté) entre les deux partitions
adjusted_rand_index = adjusted_rand_score(cluster_labels, kmeans_labels)

# Afficher les résultats
print("Centres des classes - CAH :\n", cah_centers)
print("Centres des classes - K-means :\n", kmeans_centers)
print("Silhouette Score - CAH:", silhouette_cah)
print("Silhouette Score - K-means:", silhouette_kmeans)
print("Indice de similarité (Ajusté) entre les partitions:", adjusted_rand_index)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Étape 1 : Importer les données dans Python
data = pd.read_csv('spotify_dataset.csv')

# Étape 2 : Créer un jeu de données avec des variables quantitatives spécifiques
selected_variables = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                      'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'popularity']
quantitative_data = data[selected_variables].head(500)  # Sélection des 500 premiers échantillons

# Étape 4 : Réaliser une CAH sur ces 100 classes et estimer le nombre de classes
plt.figure(figsize=(15, 10))
dendrogram(linkage(quantitative_data, method='ward'))
plt.title('Dendrogramme')
plt.xlabel('Individus')
plt.ylabel('Distance euclidienne')
plt.show()
```
